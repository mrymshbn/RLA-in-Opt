```python
import numpy as np

def linear_time_svd(A_sparse, c, k):
    d, n = A_sparse.shape
    B = np.zeros((d, c))
    
    for i in range(n):
        pi = np.linalg.norm(A_sparse[:, i], ord='fro') ** 2 / (k * np.linalg.norm(A_sparse, ord='fro') ** 2)
        for j in range(c):
            if np.random.rand() < pi:
                B[:, j] += A_sparse[:, i] / (np.sqrt(c * pi))
    
    BT = np.dot(B, B.T)
    U, S, VT = np.linalg.svd(BT)
    U = U[:, :k]
    S = np.diag(np.sqrt(S[:k]))
    VT = VT[:k, :]
    H = np.dot(np.dot(B, VT.T), np.linalg.inv(S))
    
    return H

# تعریف ماتریس A
#A = np.array([[1, 2, 3],
              #[4, 5, 6]])

c = 2  # تعداد ستون‌های ماتریس B
k = 2  # تعداد برتری‌های تک تک بردارهای ستونی A

# اجرای الگوریتم
result_H = linear_time_svd(A_sparse, c, k)
print(result_H)
```

```python
import numpy as np

def svd_algorithm(A, c, k):
    d, n = A.shape
    
    # مقداردهی اولیه به ماتریس B
    B = np.zeros((d, c))
    
    for i in range(n):
        # محاسبه احتمال p_i
        p_i = np.linalg.norm(A[:, i])**2 / np.linalg.norm(A)**2
        
        for j in range(c):
            # درج و مقیاس بندی مجدد
            B[:, j] += (p_i**0.5) * A[:, i]
            B[:, j] /= np.linalg.norm(B[:, j])
    
    # محاسبه B^T B
    B_transpose_B = np.dot(B.T, B)
    
    # محاسبه SVD از B^T B
    _, Sigma, Vt = np.linalg.svd(B_transpose_B)
    
    # انتخاب k ستون اول از H
    H_k = np.dot(B, Vt[:k, :].T) / Sigma[:k]
    
    return H_k

# تعریف یک ماتریس مثالی
A_full = np.random.rand(5, 3)
A = A_full.T  # تراپوزه ماتریس

# تعیین مقدار c و k
c = 5
k = 2

# اعمال الگوریتم
result_H_k = svd_algorithm(A, c, k)

print(f"ماتریس H_k:\n{result_H_k}")
```
