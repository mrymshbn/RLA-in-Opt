```python
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

# ایجاد یک تصویر سفید با ابعاد 800x400 پیکسل
width, height = 800, 400
image = Image.new('RGB', (width, height), color=(255, 255, 255))

# ایجاد یک شیء برای رسم روی تصویر
draw = ImageDraw.Draw(image)

# تعریف فونت و اندازه آن (باید مطمئن باشید که این فونت در سیستم شما موجود است)
try:
    font = ImageFont.truetype("arial.ttf", 30)
except IOError:
    font = ImageFont.load_default()

# متن مورد نظر
text = "Using Randomized Linear Algebra in Optimization"

# محاسبه موقعیت متن برای قرار گرفتن در مرکز تصویر
bbox = draw.textbbox((0, 0), text, font=font)
text_width = bbox[2] - bbox[0]
text_height = bbox[3] - bbox[1]
position = ((width - text_width) // 1, (height - text_height) // 3)

# رسم متن روی تصویر
draw.text(position, text, fill="blue", font=font)

# نمایش تصویر در نوت‌بوک
display(image)
```
